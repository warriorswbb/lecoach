```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
densities = [1e2, 1e3, 1e4]  # kg/m³
mass_solar = np.logspace(np.log10(0.05), np.log10(200), 100)

# Convert solar masses and radii to SI units if necessary
# Here, for simplicity, we assume R in RSun and M in MSun

plt.figure(figsize=(10, 7))

for rho in densities:
    radius = ( (3 * mass_solar) / (4 * np.pi * rho) )**(1/3)
    plt.loglog(radius, mass_solar, label=f'ρ = {rho:.0e} kg/m³')

# Observed data
observed_mass = [100, 35, 18, 6.5, 3.2, 2.1, 1.7, 1.3, 1.1, 1, 0.93, 0.85, 0.74, 0.51, 0.18, 0.11, 0.09]
observed_radius = [12, 9.8, 7.4, 3.8, 2.5, 1.7, 1.3, 1.2, 1.05, 1, 0.93, 0.85, 0.74, 0.51, 0.18, 0.11, 0.09]

plt.scatter(observed_radius, observed_mass, color='black', label='Observed Stars')

plt.xlabel('Radius (RSun)')
plt.ylabel('Mass (MSun)')
plt.title('Mass vs. Radius for Constant Density Spheres and Main-Sequence Stars')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```
